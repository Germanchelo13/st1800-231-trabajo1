# Conexión desde s3

Se leen los datos crudes que estan en la zona raw, temperatura 

```pyspark
df=spark.read.csv("s3://trabajo-1/raw/temperatura/*.csv",inferSchema=True,header=True)
```

```pyspark
#printSchema
df.printSchema()
```

```pyspark
#fisrt few rows of dataframe
df.select("Departamento", "ValorObservado").show(5)
```

```pyspark

spark.sql("""SELECT replace(temp.municipio, '"', '') as municipio,
	temp.ano as ano,
	avg(temp.valorobservado) as presion_atmosferica
FROM datos_atmosfericos.trusted_presion_atmosferica as temp
WHERE temp.ano = 2021
group by temp.ano, municipio
order by presion_atmosferica desc""").show(10)
```

```pyspark

spark.sql("""SELECT replace(temp.departamento, '"', '') as departamento,
	temp.ano as ano,
	avg(temp.valorobservado) as temperatura
FROM datos_atmosfericos.trusted_temperatura as temp
WHERE ano = 2021
group by temp.ano, temp.departamento
order by temperatura desc""").show(10)
```
